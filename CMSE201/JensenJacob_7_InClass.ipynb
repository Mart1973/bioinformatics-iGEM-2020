We are going to use NumPy to read data in from files and look at the data. The standard method for doing this in NumPy is `loadtxt`. ### Statistics
___

Now that you have read in the data, use NumPy's [statistics operations](https://docs.scipy.org/doc/numpy-1.13.0/reference/routines.statistics.html) from the pre-class to compare various properties of the water levels for all of the lakes.
* mean
* median
* standard deviation


```python
# Put your code here to print the statistical properties of each lake's water levels.
ont_mean = ont_level.mean()
ont_median = np.median(ont_level)
ont_std = ont_level.std()
print(' Ont level mean: {:.3f}\n Ont level median: {:.3f}\n Ont level STD: {:.3f}\n'.format(ont_mean, ont_median, ont_std))

mhu_mean = mhu_level.mean()
mhu_median = np.median(mhu_level)
mhu_std = mhu_level.std()
print(' MHu level mean: {:.3f}\n MHu level median: {:.3f}\n MHu level STD: {:.3f}\n'.format(mhu_mean, mhu_median, mhu_std))

eri_mean = eri_level.mean()
eri_median = np.median(eri_level)
eri_std = eri_level.std()
print(' Eri level mean: {:.3f}\n Eri level median: {:.3f}\n Eri level STD: {:.3f}\n'.format(eri_mean, eri_median, eri_std))

sup_mean = sup_level.mean()
sup_median = np.median(sup_level)
sup_std = sup_level.std()
print(' Sup level mean: {:.3f}\n Sup level median: {:.3f}\n Sup level STD: {:.3f}\n'.format(sup_mean, sup_median, sup_std))
```

```python
'''
This code is supposed to plot 3 different mathematical functions
using three different line styles and colors all on the same plot.
'''
%matplotlib inline
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 30)
y1 = x**2 #Changed to = as not a comparison
y2 = 0.3*x**3 - 3.5
y3 = 0.01*x**4 + 2*(x**2) - 10.0 #Changed to be ** to raise to power

plt.plot(x, y1, color="cornflowerblue", linestyle=":")
plt.plot(x, y2, color="salmon", marker='d', linestyle="-") #Added " to salmon to define string
plt.plot(x, y3, color="seagreen", linestyle="--", lw="2")
plt.grid() #Changed to () due to function call
```

```python
'''
This code is intended to model the approximate trajectory
of a cannonball as a function of time. The result should
look like an upside down parabola when we look at it in
the x-y domain.
'''

# Create lists for storing values
x = []
y = []
xvel = []
yvel = []

# Define gravitational acceleration
gravity = -9.8

# Define the initial conditions
# and add them to the lists
current_x = 0
current_y = 0
current_xvel = 50
current_yvel = 50

# Define the time range for the problem
# As well as the delta_t
time = np.arange(0, 10, 0.1)
delta_t = time[1] - time[0]

# Set up a loop for iterating through time
# so that we can move the cannonball forward
# in time (and consequently, space)
for t in time:
    x.append(current_x)
    y.append(current_y)
    xvel.append(current_xvel)
    yvel.append(current_yvel)
    new_x = current_x + current_xvel * delta_t
    new_y = current_y + current_yvel * delta_t
    new_xvel = current_xvel
    new_yvel = current_yvel + gravity * delta_t
    current_x = new_x
    current_y = new_y #Created storage of new y value
    current_xvel = new_xvel
    current_yvel = new_yvel

# Make plots
# Top plot: x vs. time
# Middle plot: y vs. time
# Bottom plot: y vs. x
fig, ax = plt.subplots(3,1)
fig.subplots_adjust(hspace=0.75)

ax[0].plot(time, x) #Changed to .plot
ax[0].set_xlabel("time") #Changed to set_xlabel()
ax[0].set_ylabel("x") #Changed to set_ylabel()
ax[1].plot(time, y) #Changed to .plot
ax[1].set_xlabel("time") #Changed to set_xlabel()
ax[1].set_ylabel("y") #Changed to set_ylabel()
ax[2].plot(x,y) #Changed to .plot
ax[2].set_xlabel("x") #Changed to set_xlabel()
ax[2].set_ylabel("y") #Changed to set_ylabel()
```
