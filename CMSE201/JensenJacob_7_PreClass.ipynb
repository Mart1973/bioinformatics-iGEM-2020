#### Array operations
___

The arrays look a lot like lists. But, a key difference is that you **cannot** have mixed types inside of the array. For example, try this code:

```python
my_list = [1, 3, 3] # this list has three different types in it: integer, float and string
conv_to_array = np.array(my_list) # this converts a list to an array
print(conv_to_array)
```

What are the types of the elements in the new array? Are they the same as the original list? Are they all the same as each other? Try modifying the list with different initial variables types to see if you can figure out the rule Python uses for setting the element type in the array when the conversion step happens.

The new type of all the values in the array is a string. This is not the same as the original list. The rule is that if a string exists within the array, all other types are converted into a string, regardless if that original string can be converted into a float or integer. If no string exists but at least one number is of the 'float' type, all integers are converted into floats. If no floats or strings exist in the array, the values stay as the integer type.

What NumPy arrays do is provide fast mathematical operations on one data type (e.g., floats), whereas lists are a more general container for information for which mathematical operations aren't useful.
