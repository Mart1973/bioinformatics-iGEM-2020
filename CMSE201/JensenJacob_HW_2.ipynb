# Part 2: Exploring a fundamental theorem in probability/statistics

The **central limit theorem** establishes that when independent random variables are combined, their average value tends towards a normal distribution (informally a "bell curve") even if the original variables themselves are not normally distributed. The theorem is a key concept in probability theory because it implies that probabilistic and statistical methods that work for normal distributions can be applicable to many problems involving other types of distributions.

The central limit theorem offers an explaination for the common appearance of the "bell curve" in the real world data. In cases like electronic noise, exam grades, and so on, we can often regard a single measured value as the (weighted) average of many small effects. The central limit theorem then tells us that this would often (though not always) produce a final distribution that is approximately shaped like a bell curve.

**Question 11:** Reusing the code above, write a custom function that can randomly draw a sample of numbers from your original list (of 10,000 numbers) multiple times and return a list containing the average of each random draw.

The function (named `rand_draw_averages`) needs to take as input:
1. A list of numbers (use the variable `input_list`),
2. An integer corresponding to the number of values to randomly draw from the input list (use the variable `draw_size`), and
3. An integer corresponding to the number of times to repeat the random draws (use the variable `num_draws`).

The function needs to reuturn a list called `draw_avgs` whose length should be the value provided to the variable `num_draws`. Each element in the `draw_avgs` list should be the average of each of the random draws.

```python
def rand_draw_averages (input_list = [1,2,3], draw_size = 2, num_draws = 100):
    
    draw_avgs = []
    
    #Repeats drawing until there has been 'num_draws' done
    for draw in range(num_draws):
        
        #Takes 'draw_size' samples from the input list at random and stores into a new list
        rand_draw = sample(input_list, draw_size)
        
        #Converts the list to an array to calculate mean of the draw
        randDrawAr = np.array(rand_draw)
        
        #Stores mean of array into variable
        avgAr = randDrawAr.mean()
        
        #Appends average to the list
        draw_avgs.append(avgAr)
    
    #Return the list of averages from the samples
    return draw_avgs
```
